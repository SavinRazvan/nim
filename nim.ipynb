### Train Two AIs to Prepare for an AI vs AI Game

```python
# Import necessary modules and functions from nim.py
from nim import Nim, train

# Import other necessary modules
import random

# Global variables for range settings
NUM_PILES_MIN = 3
NUM_PILES_MAX = 10
PILE_VALUE_MIN = 1
PILE_VALUE_MAX = 10

def generate_random_piles():
    """
    Generate a random number of piles with random values based on global settings.
    """
    num_piles = random.randint(NUM_PILES_MIN, NUM_PILES_MAX)
    piles = [random.randint(PILE_VALUE_MIN, PILE_VALUE_MAX) for _ in range(num_piles)]
    return piles

# Train two separate AI agents
ai1 = train(10000)
ai2 = train(10000)

def play_ai_vs_ai(ai1, ai2):
    """
    Make two AI agents play against each other and print their moves and results.
    """
    # Generate random initial piles for each game
    initial_piles = generate_random_piles()
    
    game = Nim(initial=initial_piles)
    
    # Dictionary to map player number to AI agent and AI name
    ai_dict = {0: (ai1, "AI-1"), 1: (ai2, "AI-2")}
    
    # Game loop
    while True:
        print()
        print("Piles:")
        for i, pile in enumerate(game.piles):
            print(f"Pile {i}: {pile}")
        print()
        
        # Current player
        current_player = game.player
        
        # AI makes a move
        ai, ai_name = ai_dict[current_player]
        pile, count = ai.choose_action(game.piles, epsilon=False)
        print(f"{ai_name} chose to take {count} from pile {pile}.")
        
        # Make the move
        game.move((pile, count))
        
        # Check for winner
        if game.winner is not None:
            print()
            print("GAME OVER")
            winner = ai_dict[Nim.other_player(game.winner)][1]
            print(f"Winner is {winner}")
            return
```

### AI vs AI Results:

```python
# Play the game between two AI agents
play_ai_vs_ai(ai1, ai2)
```
